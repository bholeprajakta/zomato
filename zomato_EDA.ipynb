```python
df=df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis=1)
df.head()
```

## Dropping duplicates

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.head()
```

## cleaning rate column

```python
df['rate'].unique()
```

```python
#removing new and -

def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
df['rate']=df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df.rate.isnull().sum()
```

```python
#filling null values in rate colum with mean
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isnull().sum()

# there is no null values so rate column is cleaned
```

```python
df.info()
```

## dropping null values

```python
df.dropna(inplace=True)
df.shape
```

```python
df.rename(columns={'approx_cost(for two people)':'cost2plates','listed_in(type)':'type','listed_in(city)':'city'},inplace=True)
df.columns
```

```python
df['location'].unique()
```

```python
df['city'].unique()
```

city and location both are there, lets keep only one so drop the city column

```python
df=df.drop(['city'],axis=1)
```

```python
df.head()
```

## cleaning cost2plates column

```python
df['cost2plates'].unique()
```

here cost is string, but actual cost is always integer so we remove the ,

```python
def handlecomma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
    
df['cost2plates']=df['cost2plates'].apply(handlecomma)
df['cost2plates']
```

```python
df['cost2plates'].unique()
#cost2plates column is cleaned
```

```python
df.head()
```

## cleaning rest type

```python
df['rest_type'].value_counts()
```

```python
rest_type=df['rest_type'].value_counts(ascending=False)
rest_type
```

```python
rest_type_lessthan1000=rest_type[rest_type<1000]
rest_type_lessthan1000
```

```python
rest_type_lessthan1000.count()
# 85 resto has less than 1000 rest_type
```

making rest types less than 1000 in frequency  as others

```python
def handle_rest_type(value):
    if(value in rest_type_lessthan1000):
        return 'others'
    else:
        return value
    
df['rest_type']=df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

cleaned rest_type column

## cleaning location column

```python
df['location'].value_counts()
```

```python
location=df['location'].value_counts(ascending=False)
location
```

```python
location_lessthan300=location[location<300]
location_lessthan300.count()
```

```python
def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
    
df['location']=df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

location column cleaned

## cleaning cuisines column

```python
cuisines=df['cuisines'].value_counts(ascending=False)
cuisines
```

```python
cuisine_lessthan100=cuisines[cuisines<100]
cuisine_lessthan100.count()
```

```python
def handle_cuisine(value):
    if(value in cuisine_lessthan100):
        return 'others'
    else:
        return value
    
df['cuisines']=df['cuisines'].apply(handle_cuisine)
df['cuisines'].value_counts()
```
